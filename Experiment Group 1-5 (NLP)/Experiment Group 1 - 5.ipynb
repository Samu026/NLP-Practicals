```python
# Morphological Analysis using Add-Delete Table

# Import necessary libraries
import pandas as pd
```

```python
# Sample base words
base_words = ['happy', 'kind', 'run', 'play', 'write', 'move', 'clear']
```

```python
# Prefixes and suffixes
prefixes = ['un', 're', 'dis']
suffixes = ['ing', 'ed', 'ly']
```

```python
# Create Add-Delete Table
add_delete_data = []

for word in base_words:
    # Add Prefixes
    for prefix in prefixes:
        new_word = prefix + word
        add_delete_data.append({
            'Operation': 'Add',
            'Morpheme': prefix,
            'Base Word': word,
            'Result Word': new_word
        })
        # Delete Prefix
        if new_word.startswith(prefix):
            base = new_word[len(prefix):]
            add_delete_data.append({
                'Operation': 'Delete',
                'Morpheme': prefix,
                'Base Word': new_word,
                'Result Word': base
            })

    # Add Suffixes
    for suffix in suffixes:
        new_word = word + suffix
        add_delete_data.append({
            'Operation': 'Add',
            'Morpheme': suffix,
            'Base Word': word,
            'Result Word': new_word
        })
        # Delete Suffix
        if new_word.endswith(suffix):
            base = new_word[:-len(suffix)]
            add_delete_data.append({
                'Operation': 'Delete',
                'Morpheme': suffix,
                'Base Word': new_word,
                'Result Word': base
            })
```

```python
# Create DataFrame
add_delete_df = pd.DataFrame(add_delete_data)
```
